```python
import pandas
import numpy

import matplotlib.pyplot as plt
```

```python
fl = "/Users/simondi/PROJECTS/target_infect_x_project/src/tix-analysis/data/factor_analysis_factors.tsv"
```

```python
df = pandas.read_csv(fl, sep="\t").values.transpose()
```

```python
var = numpy.apply_along_axis(lambda x: sum(x ** 2) / 15 , 0, df)
```

```python
s = numpy.cumsum(var)
```

```python
s
```

```python
plt.bar(range(15), s)
```

```python
plt.style.use(["seaborn-whitegrid"])
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Dasd']

_, ax = plt.subplots(figsize=(7, 4), dpi=720)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.xaxis.set_label_coords(x=.9, y=-0.1)
ax.yaxis.set_label_coords(x=-0.05, y=.70)
ax.grid(linestyle="")
ax.grid(which="major", axis="y", linestyle="-", color="gainsboro")
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
bar = plt.bar(list(map(str, range(1, 16))), s, color="#696969")
for i, rect in enumerate(bar):
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, s[i], '%s' % str(round(s[i], 2)), ha='center', va='bottom')

plt.xlabel('# factors', fontsize=15)
plt.ylabel("Cumulative Variance", fontsize=15)
plt.show()
```
